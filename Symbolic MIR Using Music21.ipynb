## Markov chains

```python
c = [0.2, 0.3, 0.5]
d = [0.5, 0, 0.5]
e = [0.7, 0.1, 0.2]

mc = [c,d,e]
```

```python

def markov_chain(mc, num, start= 0):
    cur_note = start
    chain = [start]
    for n in range(num-1):
        prob = random.random()
        for i, p in enumerate(mc[cur_note]):
            prob -= p
            if prob < 0:
                break
        chain.append(i)
    return chain
            
```

```python
markov_chain(mc, 10)
```

```python
notes = [0,2,4]
out = [notes[i] for i in markov_chain(mc, 10)]
```
